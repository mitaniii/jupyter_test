# seaborn サンプル

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import japanize_matplotlib
```

```python
train = pd.read_csv('titanic3.csv')

sns.set(style="darkgrid", font="IPAEXGothic")
g = sns.barplot(x = 'sex', y = 'survived', data=train, palette='Set2')
g.set_title('男女の生存率')
g.set_xlabel('性別')
g.set_ylabel('生存率')
```

```python
sns.countplot(x=train['pclass'], palette='Set3')
```

```python
sns.displot(train['age'], bins=8, color='red')
```
