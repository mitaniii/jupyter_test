## matplotlib

```python
import matplotlib.pyplot as plt #グラフのライブラリ
import japanize_matplotlib
import os
output_dir = '..data/output' #出力先のディレクトリを設定
```

```python
plt.figure(figsize=(3, 3)) #表のサイズ指定
plt.plot([1, 2, 3, 4], [1, 4, 9, 16]) #折れ線グラフ(x, y)
plt.plot([1, 2, 3, 4], [2, 6, 10, 17], ":", c='green', alpha=0.5, linewidth=3) #2本目(オプションいろいろ)
```

```python
plt.figure(figsize=(3, 3))
x = [1, 2, 3, 4]
plt.plot(x, [1, 4, 9, 16], '-', label='y=x^2', marker='o')
plt.plot(x, [2, 4, 6, 12], '--', label='y=x^2', marker='v')
plt.plot(x, [3, 5, 14, 8], '-.', label='y=x^2', marker='.')
plt.plot(x, [4, 6, 10, 18], ':', label='y=x^2', marker='h')
plt.legend()
```

```python
plt.figure(figsize=(3, 3))
x = [1, 2, 3, 4]
y = [200, 400, 200, 500]
label = ['A', 'B', 'C', 'D']
plt.bar(x, y, tick_label=label) #棒グラフ
```

```python
plt.figure(figsize=(3, 3))

plt.scatter([1, 2, 3, 4], [1, 4, 9, 16], label='y=x^2', marker='o')
plt.legend()
```

```python
import numpy as np

x = np.random.rand(30) #ランダムに30個
y = np.random.rand(30)

plt.figure(figsize=(3, 3))
plt.scatter(x, y, s=20, c='red', alpha=0.5) #散布図
```

```python
label = ['A', 'B', 'C', 'D', 'E', 'F']
x = [np.random.randint(1, 10) for _ in range(len(label))]

plt.figure(figsize=(3, 3))
plt.pie(x, labels=label, counterclock=False, startangle=90) #円グラフ
```

```python
fig, ax = plt.subplots(2,2, figsize=(3, 3))
ax[0][0].plot([1, 2, 3, 4], [1, 4, 9, 14]) #折れ線グラフ
ax[0][0].set_title('折れ線グラフ')

ax[1][1].pie([50, 30, 15, 5],labels=['A','B','C','D'])
```
